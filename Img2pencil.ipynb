```python
# Importing the necessary modules

import cv2
import matplotlib.pyplot as plt
```

```python
# Reading the image

img = cv2.imread('birds.jpg')
```

```python
# Plotting the image

img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
# Converting the image to Grayscale image

img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap = 'gray')
plt.show()
```

```python
# Inverting the image

img_invert = cv2.bitwise_not(img_gray)
plt.imshow(img_invert, cmap = 'gray')
plt.show()
```

```python
# Smoothing the image

img_smooth = cv2.GaussianBlur(img_invert,(21,21),0)
plt.imshow(img_smooth, cmap = 'gray')
plt.show()
```

```python
# Converting image to pencil sketches

sketch = cv2.divide(img_gray, 255-img_smooth, scale=255)
plt.imshow(sketch, cmap = 'gray')
plt.show()
```
