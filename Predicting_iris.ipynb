```python
# importing the necessary library

import pandas as pd 
from sklearn import tree
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading our dataset using read_csv method of pandas

iris = pd.read_csv('iris.csv')
```

```python
# Here we have printed five rows using head method

iris.head()
```

```python
# Finding the number of rows and columns in the dataset

iris.shape
```

```python
# we have printed the datatypes and checked that is there any null values present in the dataset or not.

iris.info()
```

```python
# Dropping the ID column from our dataset as it does not require

iris.drop('Id', axis=1, inplace = True)
```

```python
iris.sample(5)
```

```python
# Spliting dataset into input and target variables

input = iris.drop('Species', axis=1)
target = iris['Species']
```

```python
dtc = tree.DecisionTreeClassifier()
```

```python
# Here we have fitted our data to the Decision Tree Classifier model

dtc.fit(input, target)
```

```python
#Finding the score of our model

dtc.score(input, target)
```

```python
# Plotting the tree graph and visualizing graphically

feature = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
clas = ['Iris-setosa','Iris-versicolor','Iris-virginica']

plt.figure(figsize=(20,10))
tree.plot_tree(dtc,feature_names=feature,class_names=clas, filled=True)
plt.show()
```

```python
# Predicting

dtc.predict([[5.1,3.5,1.4,0.2]])
```
